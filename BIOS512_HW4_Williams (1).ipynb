```r
full_long <- full %>%
  pivot_longer(
    cols = -c(patient_id, name), 
    names_to = "variable", 
    values_to = "value", 
    values_transform = list(value = as.character)
  )

obs_per_patient <- full_long %>%
  group_by(patient_id, name) %>% 
  tally(name = "n_observations") %>% 
  arrange(desc(n_observations))

obs_per_patient
```

Question 4

```r
full_chr <- full %>%
  select(patient_id, name, where(is.character)) %>%
  pivot_longer(
    cols = -c(patient_id, name),
    names_to = "variable_chr",
    values_to = "value_chr"
  ) %>%
  group_by(patient_id, name) %>%
  mutate(row = row_number()) %>%
  ungroup()

full_num <- full %>%
  select(patient_id, name, where(is.numeric)) %>%
  pivot_longer(
    cols = -c(patient_id, name),
    names_to = "variable_num",
    values_to = "value_num"
  ) %>%
  group_by(patient_id, name) %>%
  mutate(row = row_number()) %>%
  ungroup()

full_date <- full %>%
  select(patient_id, name, where(lubridate::is.Date)) %>%
  pivot_longer(
    cols = -c(patient_id, name),
    names_to = "variable_date",
    values_to = "value_date"
  ) %>%
  group_by(patient_id, name) %>%
  mutate(row = row_number()) %>%
  ungroup()

full_long_types <- full_chr %>%
  full_join(full_num, by = c("patient_id", "name", "row")) %>%
  full_join(full_date, by = c("patient_id", "name", "row"))

obs_type_per_patient <- full_long_types %>%
  group_by(patient_id, name) %>%
  summarise(
    n_chr  = sum(!is.na(value_chr)),
    n_num  = sum(!is.na(value_num)),
    n_date = sum(!is.na(value_date)),
    .groups = "drop"
  ) %>%
  arrange(desc(n_chr))

obs_type_per_patient
```

```r
 state_map <- tibble(
     state_full = state.name,   
     state_abbr = state.abb     
 )
 
 us_state_counts <- df %>%
     filter(country == "USA") %>%                
     distinct(id, .keep_all = TRUE) %>%          
     mutate(state_clean = case_when(
         state %in% state.abb ~ state,                                      
         state %in% state.name ~ state_map$state_abbr[match(state, state_map$state_full)], 
         TRUE ~ NA_character_                                              
     )) %>%
     filter(!is.na(state_clean)) %>%                                        
     group_by(state_clean) %>%
     tally(name = "num_sightings") %>%
     filter(num_sightings > 1) %>%                                         
     arrange(desc(num_sightings))
 
 
 ggplot(us_state_counts, aes(x = reorder(state_clean, -num_sightings), y = num_sightings)) +
     geom_bar(stat = "identity", fill = "steelblue") +
     labs(title = "Number of UFO Sightings by US State",
          x = "State",
          y = "Number of Sightings") +
     theme_minimal() +
     theme(axis.text.x = element_text(angle = 90, hjust = 1, vjust = 0.5))
```
