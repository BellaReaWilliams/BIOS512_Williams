one-hot coding: This method transforms unique values from a single variable into a new, separate binary variable column.

feature selection: A process of automatically identifying and selecting the most important/relevant variables in the data.

classifier: A type of model that assigns inputs to predefined categories instead of predicting continuous numerical values.

precision: A metric used to evaluate a classifier's performance, which measures the proportion of positive identifications that were actually correct.

recall: Another classifier performance metric that measures the proportion of actual positive identifications that were correctly identified by the model.

F1 score: A metric that combines a classifier's precision and recall into a unified measure, representing the harmonic mean. 

parsimonious model: A statistical model that aims to achieve good predictive performance while using the fewest possible parameters. The product will be as simple as possible while still explaining the data well.

ridge regression: A type of regression that adds a penalty proportional to the sum of the squared coefficients. 

LASSO regression: A type of regression that adds a penalty proportional to the sum of the absolute values of the coefficients. 

cross validation: A statistical technique used to assess how well a model will generalize to an independent dataset and reliably compare the performance of different models. 

tree based methods: A category of machine learning model that uses a decision-tree structure to make predictions, offering an alternative to linear models (regression).

A)

```r
d <- read_csv("diabetes.csv")
```

```r
set.seed(123)
```

```r
names(d) <- names(d) %>%
  str_replace_all("-","_") %>%
  str_replace_all("[^_A-Za-z0-9]*","") %>%
  str_to_lower()
```

```r
cat("--- Structure of the Data ---\n")
str(d)
```

outcome is a binary, categorical variable so a logistic regression would be appropriate.

B)

```r
scale <- function(a){
  (a - min(a)) / (max(a) - min(a))
}

d_num <- d %>%
  mutate(across(where(is.numeric) & !matches("outcome"), scale))

train <- runif(nrow(d_num)) < 0.75
test <- !train
```

C)

```r
f = outcome ~ pregnancies + glucose + bloodpressure + skinthickness +
    insulin + bmi + diabetespedigreefunction + age

m <- glm(f, data = d_num %>% filter(train), family = binomial())
summary(m)
```

D)

```r
dx <- d_num %>% filter(test)

dx <- dx %>%
  mutate(prob = predict(m, dx %>% as.data.frame(), type = "response"),
         pred = ifelse(prob > 0.5, 1, 0))

TP <- sum(dx$pred == 1 & dx$outcome == 1)
FP <- sum(dx$pred == 1 & dx$outcome == 0)
TN <- sum(dx$pred == 0 & dx$outcome == 0)
FN <- sum(dx$pred == 0 & dx$outcome == 1)

acc <- (TP + TN) / (TP + TN + FP + FN)

tibble(
  True_Positive = TP,
  False_Positive = FP,
  True_Negative = TN,
  False_Negative = FN,
  Accuracy = acc
)
```

E)

```r
install.packages("glmnet")
```

```r
library(glmnet)

set.seed(123)

x_train <- d_num %>% filter(train) %>% select(-outcome) %>% as.matrix()
y_train <- d_num %>% filter(train) %>% pull(outcome)

lasso <- cv.glmnet(x_train, y_train, family = "binomial", alpha = 1)

coef_lasso <- coef(lasso, s = "lambda.min")
coef_lasso   

x_test <- d_num %>% filter(test) %>% select(-outcome) %>% as.matrix()
y_test <- d_num %>% filter(test) %>% pull(outcome)

prob_lasso <- predict(lasso, newx = x_test, s = "lambda.min", type = "response")
pred_lasso <- ifelse(prob_lasso > 0.5, 1, 0)

TP_l <- sum(pred_lasso == 1 & y_test == 1)
FP_l <- sum(pred_lasso == 1 & y_test == 0)
TN_l <- sum(pred_lasso == 0 & y_test == 0)
FN_l <- sum(pred_lasso == 0 & y_test == 1)

acc_lasso <- (TP_l + TN_l) / (TP_l + TN_l + FP_l + FN_l)
acc_lasso 
```

F)

```r
plot_df <- tibble(
  actual = y_test,
  predicted = as.numeric(prob_lasso)
)

ggplot(plot_df, aes(x = actual, y = predicted)) +
  geom_jitter(width = 0.1, height = 0) +
  geom_smooth(method = "lm", se = FALSE) +
  labs(title = "Actual vs Predicted (Lasso Logistic Model)",
       x = "Actual Outcome (0/1)",
       y = "Predicted Probability")
```
